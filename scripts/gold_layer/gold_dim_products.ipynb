```python
%sql
use lakehouse.silver
```

```python
query = '''

select distinct sp.product_id,
                sp.product_name, 
                sp.category, 
                so.payment_method,
                so.status

from silver.products sp join silver.orders so 
on sp.product_id = so.product_id       

'''

df = spark.sql(query)
```

```python
df.display()
```

```python
df.write.format('delta').mode('overwrite').saveAsTable('lakehouse.gold.dim_products')
```
