```python
%sql
use lakehouse.silver
```

```python
query = '''
 
 select so.order_id,
        dense_rank()over(order by so.customer_id) as customer_sk,
        so.product_id,
        so.order_amount,
        so.quantity,
        so.order_date,
        sp.price as product_price

from silver.orders so join silver.products sp
on so.product_id = sp.product_id

'''

df = spark.sql(query)
```

```python
df.display()
```

```python
df.write.format('delta').mode('overwrite').saveAsTable('lakehouse.gold.fact_orders')
```
